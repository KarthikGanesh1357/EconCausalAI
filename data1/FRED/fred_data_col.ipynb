```python
import requests
import pandas as pd
import time
import string

API_KEY = "351c428d6e217816be5e0f408666ed89"
BASE_URL = "https://api.stlouisfed.org/fred/series/search"

limit = 1000
offset = 0
all_series = []

# Search all letters a-z to capture maximum variety of series
for letter in string.ascii_lowercase:
    offset = 0
    print(f"Searching with letter: {letter}")
    while True:
        params = {
            "api_key": API_KEY,
            "search_text": letter,
            "file_type": "json",
            "limit": limit,
            "offset": offset
        }

        response = requests.get(BASE_URL, params=params)
        if response.status_code != 200:
            print("Error:", response.status_code, response.text)
            break

        data = response.json()
        results = data.get("seriess", [])

        if not results:
            break

        all_series.extend(results)
        print(f"Letter {letter}: Retrieved {len(results)} records at offset {offset}")
        offset += limit
        time.sleep(0.2)  # Avoid rate limiting

# Save all results
if all_series:
    df = pd.DataFrame(all_series).drop_duplicates(subset="id")
    df.to_csv("fred_all_series_comprehensive.csv", index=False)
    print(f"Saved {len(df)} unique series to fred_all_series_comprehensive.csv")
else:
    print("No data retrieved.")
```
